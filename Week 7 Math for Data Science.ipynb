```python
import numpy as np

# Define Matrix A (3x4)
A = np.array([
    [1, -2, 3, 7],
    [2, 1, 1, 4],
    [-3, 2, -2, 10]
])

# Define Matrix B (3x4)
B = np.array([
    [0, 1, -3, -2],
    [10, -1, 2, -3], 
    [5, 1, -1, 4]
])

# Define Matrix C (4x4)
C = np.array([
    [4, 0, -2, 3],
    [3, 6, 9, 7],
    [2, 2, 5, 1],
    [9, 4, 6, -2]
])

print("--- Matrices Defined ---\n")
print("Matrix A (3x4):\n", A)
print("\nMatrix B (3x4):\n", B)
print("\nMatrix C (4x4):\n", C)

# --- Question 2: Display Matrix Dimensions ---
print("\n--- Matrix Dimensions ---")
print(f"Dimensions of A (A.shape): {A.shape}")
print(f"Dimensions of B (B.shape): {B.shape}")
print(f"Dimensions of C (C.shape): {C.shape}")
```

### 3. Determine if Operations are Defined

#### (a) A + B
**Defined.** Matrix addition requires identical dimensions. Since Matrix A is $3 \times 4$ and Matrix B is $3 \times 4$, the sum is defined.

#### (b) A B
**Not Defined.** Matrix multiplication $A \times B$ requires the number of columns in $A$ (4) to equal the number of rows in $B$ (3). Since $4 \ne 3$, $A B$ is **not possible**.

#### (c) A C
**Defined.** Matrix multiplication $A \times C$ requires the number of columns in $A$ (4) to equal the number of rows in $C$ (4). Since $4 = 4$, $A C$ is defined. The result will be a $3 \times 4$ matrix.

#### (d) C^T
**Defined.** The transpose operation is always defined for any matrix.

```python
# --- Question 3: Calculations ---

# (a) A + B
A_plus_B = A + B
print("--- Result of A + B ---\n", A_plus_B)

# (c) A C
A_times_C = np.dot(A, C)
print("\n--- Result of A C ---\n", A_times_C)

# (d) C^T
C_transpose = C.T
print("\n--- Result of C^T ---\n", C_transpose)
```

### 4. Illustrate the Property: $(A+B)^T = A^T + B^T$

The property is illustrated by calculating both sides of the equation and showing they produce the identical result:

* **Left Hand Side (LHS):** $(A+B)^T$
* **Right Hand Side (RHS):** $A^T + B^T$

As verified by Python, the two resulting matrices are numerically equal, confirming the property for these matrices.

```python
# --- Question 4: Illustrate (A+B)^T = A^T + B^T ---

# Left Hand Side (LHS): (A+B)^T
LHS_4 = (A + B).T

# Right Hand Side (RHS): A^T + B^T
RHS_4 = A.T + B.T

print("--- Property Illustration: (A+B)^T = A^T + B^T ---")
print("LHS: (A+B)^T\n", LHS_4)
print("\nRHS: A^T + B^T\n", RHS_4)

# Check if they are numerically equal
print("\nLHS and RHS are equal:", np.array_equal(LHS_4, RHS_4))
```

### 5. Illustrate the Property: $(AC)^T = C^T A^T$

This property states that the transpose of a product is the product of the transposes in reverse order. The illustration confirms this:

* **Left Hand Side (LHS):** $(AC)^T$
* **Right Hand Side (RHS):** $C^T A^T$

Both calculation paths yielded the same $4 \times 3$ matrix, thus illustrating the property.

```python
# --- Question 5: Illustrate (AC)^T = C^T A^T ---

# Left Hand Side (LHS): (AC)^T
# A_times_C was calculated in Step 3
LHS_5 = (A_times_C).T

# Right Hand Side (RHS): C^T A^T
# C.T is C transpose, A.T is A transpose. Use np.dot() for multiplication.
RHS_5 = np.dot(C.T, A.T)

print("--- Property Illustration: (AC)^T = C^T A^T ---")
print("LHS: (AC)^T\n", LHS_5)
print("\nRHS: C^T A^T\n", RHS_5)
print("\nLHS and RHS are equal:", np.array_equal(LHS_5, RHS_5))
```

```python
# --- Question 6: Find C inverse (C^-1) ---

# Use numpy.linalg.inv() to find the inverse of C
try:
    C_inverse = np.linalg.inv(C)

    print("--- Result of C^-1 (Inverse of C) ---\n")
    # Round to 4 decimal places for clarity
    print("Matrix C^-1:\n", C_inverse.round(4)) 

    # Verification: C * C^-1 should be the Identity Matrix (I)
    print("\n--- Verification: C * C^-1 (should be Identity Matrix) ---\n")
    # Use np.allclose for reliable comparison of floating-point numbers
    print(np.allclose(np.dot(C, C_inverse), np.identity(4))) 

except np.linalg.LinAlgError:
    print("Matrix C is singular and its inverse cannot be calculated.")
```

### 6. Find $C^{-1}$ using Python code.
